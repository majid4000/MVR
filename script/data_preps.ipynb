```python
import pandas as pd
from tqdm.notebook import tqdm
import json
import os

import config

tqdm.pandas()
```

```python
#Tweet
with open(config.tweets_x, "r") as f:
    t_x = f.read().split('\n')
with open(config.tweets_y, "r") as f:
    t_y = f.read().split('\n')

df_tw = pd.DataFrame(
    {'text': t_x,
     'label': t_y,
    })

#GoogleNews
with open(config.googel_news_x, "r") as f:
    gn_x = f.read().split('\n')
with open(config.googel_news_y, "r") as f:
    gn_y = f.read().split('\n')

df_gn = pd.DataFrame(
    {'text': gn_x,
     'label': gn_y,
    })

#StackOverflow
with open(config.stack_overflow_x, "r") as f:
    sf_x = f.read().split('\n')
with open(config.stack_overflow_y, "r") as f:
    sf_y = f.read().split('\n')

df_sf = pd.DataFrame(
    {'text': sf_x,
     'label': sf_y,
    })
```

```python

with open(config.googel_news_og, 'w' , encoding = 'utf-8') as fp:
    json.dump(df_gn.to_dict('records'), fp , ensure_ascii=False, indent=2)

with open(config.stack_overflow_og, 'w' , encoding = 'utf-8') as fp:
    json.dump(df_sf.to_dict('records'), fp , ensure_ascii=False, indent=2)

with open(config.tweets_og, 'w' , encoding = 'utf-8') as fp:
    json.dump(df_tw.to_dict('records'), fp , ensure_ascii=False, indent=2)
```
